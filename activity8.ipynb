<h1>Narcotics</h1>

```python
# My Filter (I HAVE TO CREATE MY OWN FILTER)
narcotics_filter = df[(df['Primary Type'] == "NARCOTICS") &
                      (df['Arrest'] == True)]

# AGGREGATE by latitude and longitude (I HAVE TO REPLACE THE groupedby with my FILTERED DATA)
aggregated_df = narcotics_filter.groupby(['Latitude', 'Longitude']).size().reset_index(name= 'incident_count')

# aggregate incidents by location 
aggregated_df ['normalize_weight'] = (aggregated_df['incident_count'] - aggregated_df['incident_count'].min()) / \
                                (aggregated_df['incident_count'].max() - aggregated_df['incident_count'].min())



# Drop missing coordinates just in case
aggregated_df = aggregated_df.dropna(subset=['Latitude', 'Longitude'])

# Convert to list of [lat, lon]
heat_data = aggregated_df[['Latitude', 'Longitude','normalize_weight']].values.tolist()

# Create base map
m = folium.Map(location=[ 41.8781, -87.6298], zoom_start=10)

# Add heatmap
HeatMap(heat_data).add_to(m)

# Save map to file
m.save('narcotics_filter.html')

# Display map 
m
```
